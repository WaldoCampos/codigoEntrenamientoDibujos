```python
import subprocess
import glob
import os
import shutil
```

```python

script = "/users/btorres/proyectos/redesNeuronales/Pidinet/pidinet-master/main.py"
saveDir = "/users/btorres/databases/ecommerce/pidinet_gaussian1/"
dataDir = "/users/btorres/databases/ecommerce/ecommerce_gaussian1/"
eval = "/users/btorres/proyectos/redesNeuronales/Pidinet/pidinet-master/trained_models/table7_pidinet.pth --evaluate-converted"
```

```python
for class_path in glob.glob(os.path.join(dataDir, '**')):
    class_name = os.path.basename(class_path) + "/"
    print(class_name)

    newpath = saveDir + class_name
    if not os.path.exists(newpath[:-1]):
       os.makedirs(newpath)

    cmd = "python " + script + " --model pidinet_converted --config carv4 --sa --dil -j 4 --gpu 1 --savedir " + saveDir + class_name + " --datadir " + dataDir + class_name + " --dataset Custom --evaluate " + eval + " --evaluate-converted"

    p = subprocess.call(cmd, shell=True)

    if type(p) is not int:
        out, err = p.communicate()
        result = out.split('\n')
        for lin in result:
            if not lin.startswith('#'):
                print(lin)
```
