```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torchvision
from torchvision.io import read_image, write_jpeg
import torchvision.transforms.functional as F
import torchvision.transforms as T

import glob
import os

from skimage.io import imread, imsave, imshow
```

```python

dataDir = "/home/vision/smb-datasets/ecommerce/images/train/"
#pidinet = "/home/vision/smb-datasets/ecommerce/ecommerce_pidinet/"
saveDir = "/users/btorres/databases/ecommerce/train/"
```

```python
o = 0
for class_path in glob.glob(os.path.join(dataDir, '**')):

    class_name = os.path.basename(class_path) + "/"
    print(class_name)

    newpath = saveDir + class_name
    if not os.path.exists(newpath[:-1]):
        os.makedirs(newpath)

    for sketch_path in glob.iglob(os.path.join(class_path, '**')):
        image_name = os.path.basename(sketch_path)
        save_path = saveDir + class_name + "/" + image_name.split(".")[-2]

        if len(image_name.split("_")) > 1:
            continue
        else:
            img = read_image(sketch_path).to('cuda')


            j = 1
            for i in np.linspace(-0.5, 0.4, 10):
                new = F.adjust_hue(img, i)

                new = new.type(torch.uint8)
                resultado= new.to('cpu')


                write_jpeg(resultado, save_path + "_" + str(j) + ".jpg")
                j += 1


            if (o % 100 == 0):
                print(str(o) + " imágenes procesadas")

            o += 1
```
